```python
#proportion of sentiment category in different platform
import pandas as pd

# Reload the original dataset
file_path_original = '/Users/luomengxin/Downloads/Youtube-Feminism_Related_Comments.csv'
data_original = pd.read_csv(file_path_original)

# Reload the new dataset
file_path_new = '/Users/luomengxin/Downloads/IMBD-Feminism_Related_Comments.csv'
data_new = pd.read_csv(file_path_new)

# Calculating the proportions of each sentiment category in both datasets
sentiment_proportions_original = data_original['Sentiment Category'].value_counts(normalize=True)
sentiment_proportions_new = data_new['Sentiment Category'].value_counts(normalize=True)

sentiment_proportions_original, sentiment_proportions_new
```

```python
import matplotlib.pyplot as plt

# Creating pie charts for both datasets
fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for the original dataset (YouTube)
ax[0].pie(sentiment_proportions_original, labels=sentiment_proportions_original.index, autopct='%1.1f%%', startangle=140)
ax[0].set_title('Sentiment Category Proportions (YouTube)')

# Pie chart for the new dataset (IMDB)
ax[1].pie(sentiment_proportions_new, labels=sentiment_proportions_new.index, autopct='%1.1f%%', startangle=140)
ax[1].set_title('Sentiment Category Proportions (IMDB)')

plt.tight_layout()
plt.show()
```

```python
# Increasing font size and changing the theme color for the pie charts
plt.rcParams['font.size'] = 16  # Increasing the font size
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']  # Defining a new color palette

fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for the original dataset (YouTube) with updated font size and color theme
ax[0].pie(sentiment_proportions_original, labels=sentiment_proportions_original.index, autopct='%1.1f%%', startangle=140, colors=colors)
ax[0].set_title('Sentiment Category Proportions (YouTube)')

# Pie chart for the new dataset (IMDB) with updated font size and color theme
ax[1].pie(sentiment_proportions_new, labels=sentiment_proportions_new.index, autopct='%1.1f%%', startangle=140, colors=colors)
ax[1].set_title('Sentiment Category Proportions (IMDB)')

plt.tight_layout()
plt.show()
```

```python
# Function to calculate sentiment proportions
def calculate_sentiment_proportions(data, sentiment_column='Sentiment Category'):
    sentiment_counts = data[sentiment_column].value_counts(normalize=True)
    return sentiment_counts

# Now we will load each dataset and calculate the sentiment proportions
file_path_original_youtube = '/Users/luomengxin/Downloads/Youtube-Feminism_Related_Comments.csv'
file_path_original_imdb = '/Users/luomengxin/Downloads/IMBD-Feminism_Related_Comments.csv'
file_path_imdb_cleaned = '/Users/luomengxin/Downloads/IMDB_reviews_cleaned-Sentiment.csv'
file_path_youtube_cleaned = '/Users/luomengxin/Downloads/Youtube-final_data-Sentiment.csv'

# Read the datasets
data_original_youtube = pd.read_csv(file_path_original_youtube)
data_original_imdb = pd.read_csv(file_path_original_imdb)
data_imdb_cleaned = pd.read_csv(file_path_imdb_cleaned)
data_youtube_cleaned = pd.read_csv(file_path_youtube_cleaned)

# Calculate the sentiment proportions
sentiment_proportions_youtube_original = calculate_sentiment_proportions(data_original_youtube)
sentiment_proportions_imdb_original = calculate_sentiment_proportions(data_original_imdb)
sentiment_proportions_youtube_cleaned = calculate_sentiment_proportions(data_youtube_cleaned)
sentiment_proportions_imdb_cleaned = calculate_sentiment_proportions(data_imdb_cleaned)

# Now we will plot the pie charts
colors = ['#ff9999','#66b3ff','#99ff99']  # Colors for negative, positive, and neutral

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 14))

# Pie chart for the original YouTube dataset
axs[0, 0].pie(sentiment_proportions_youtube_original, labels=sentiment_proportions_youtube_original.index, autopct='%1.1f%%', startangle=140, colors=colors)
axs[0, 0].set_title('YouTube (Feminism-Related)')

# Pie chart for the original IMDB dataset
axs[0, 1].pie(sentiment_proportions_imdb_original, labels=sentiment_proportions_imdb_original.index, autopct='%1.1f%%', startangle=140, colors=colors)
axs[0, 1].set_title('IMDB (Feminism-Related)')

# Pie chart for the cleaned YouTube dataset
axs[1, 0].pie(sentiment_proportions_youtube_cleaned, labels=sentiment_proportions_youtube_cleaned.index, autopct='%1.1f%%', startangle=140, colors=colors)
axs[1, 0].set_title('YouTube (Original)')

# Pie chart for the cleaned IMDB dataset
axs[1, 1].pie(sentiment_proportions_imdb_cleaned, labels=sentiment_proportions_imdb_cleaned.index, autopct='%1.1f%%', startangle=140, colors=colors)
axs[1, 1].set_title('IMDB (Original)')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
