```python
import matplotlib.pyplot as plt

# Function to plot a pie chart for each column
def plot_pie_chart(data, column_name):
    # Count the frequency of each category
    sentiment_counts = data[column_name].value_counts()

    # Plot
    plt.figure(figsize=(6, 6))
    plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Sentiment Distribution in {column_name}')
    plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

# Plotting pie charts for each column
for column in data.columns:
    plot_pie_chart(data, column)

plt.show()
```

```python
# Set up the figure layout to plot all pie charts in a 2x2 grid, similar to the example provided
fig, axs = plt.subplots(2, 2, figsize=(14, 14))

# Flatten the axes array for easy iteration
axs = axs.flatten()

# Loop through each column and create pie charts in the respective subplot
for i, column in enumerate(data.columns):
    sentiment_counts = data[column].value_counts()
    axs[i].pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140)
    axs[i].set_title(f'{column}')
    axs[i].axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

# Adjust the layout to prevent overlapping titles and labels
plt.tight_layout()

# Save the figure to a file
output_file_path = '/Users/luomengxin/Downloads/combined_sentiment_pie_charts.png'
plt.savefig(output_file_path)

# Display the combined figure
plt.show()

output_file_path
```

```python
# Define the color scheme for negative, positive, and neutral sentiments
colors = ['#ff9999', '#66b3ff', '#99ff99']

# Set up the figure layout to plot all pie charts in a 2x2 grid, with specified colors and font size
fig, axs = plt.subplots(2, 2, figsize=(14, 14))
axs = axs.flatten()

# Set font size for all elements
plt.rcParams.update({'font.size': 16})

# Loop through each column and create pie charts in the respective subplot with the new settings
for i, column in enumerate(data.columns):
    sentiment_counts = data[column].value_counts()
    # Rearrange the colors array to match the order of sentiment_counts
    ordered_colors = [colors[sentiment_counts.index.get_loc(label)] for label in sentiment_counts.index]
    axs[i].pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140, colors=ordered_colors)
    axs[i].set_title(f'{column}')
    axs[i].axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

# Adjust the layout to prevent overlapping titles and labels
plt.tight_layout()

# Save the figure to a file
output_file_path_modified_colors = '/Users/luomengxin/Downloads/combined_sentiment_pie_charts_modified_colors.png'
plt.savefig(output_file_path_modified_colors)

# Display the combined figure
plt.show()

output_file_path_modified_colors
```
